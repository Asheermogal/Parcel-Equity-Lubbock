```python
df2012.info()
```

### 4

```python
# Function to append the correct suffix to addresses
def append_address_suffix(address):
    if "TX" not in address and "Lubbock" not in address:
        # Append ", Lubbock, TX" if neither "TX" nor "Lubbock" is present
        return address + ", Lubbock, TX"
    elif "TX" not in address:
        # Append ", TX" if "TX" is missing
        return address + ", TX"
    else:
        # If both are present, return the address as-is
        return address

# Apply the function to the 'Situs' column
df2012.loc[:, 'Situs'] = df2012['Situs'].apply(append_address_suffix)
```

```python
# Function to clean cells by taking the first value if the list has only one element
def clean_single_value_lists(cell):
    if isinstance(cell, list) and len(cell) == 1:  # Check if it's a list with only one value
        return cell[0]  # Extract the single value
    return cell  # Leave as is if it's not a single-value list

# Apply the function across all columns
final_df = final_df.apply(lambda col: col.map(clean_single_value_lists))
```

```python
# Function to get geocoding details
def get_geocoding_details(address, client):
    try:
        geocode_result = client.geocode(address)
        if geocode_result:
            # Extract required details
            formatted_address = geocode_result[0]['formatted_address']
            location = geocode_result[0]['geometry']['location']
            latitude = location['lat']
            longitude = location['lng']
            return formatted_address, latitude, longitude
        else:
            return None, None, None
    except Exception as e:
        print(f"Error geocoding {address}: {e}")
        return None, None, None

# Apply geocoding to the dataframe
def batch_geocode(df, client, batch_size, max_records):
    processed_rows = 0
    results = []
    for idx, address in enumerate(df['Situs']):
        if processed_rows >= max_records:  # Stop after max_records
            break
        # Print progress
        print(f"\n{'='*40}")
        print(f"Processing Address {processed_rows + 1}/{max_records}")
        print(f"Address: {address}")

        # Get geocoding details
        formatted_address, latitude, longitude = get_geocoding_details(address, client)

        # Print the response
        print(f"Response: Standard Address: {formatted_address}, Latitude: {latitude}, Longitude: {longitude}")

        # Append results
        results.append((formatted_address, latitude, longitude))
        processed_rows += 1

        # Pause for rate limit every batch_size addresses
        if idx > 0 and idx % batch_size == 0:
            print(f"\n{'-'*40}")
            print(f"Batch of {batch_size} processed. Pausing for 60 seconds...")
            time.sleep(60)
            print(f"Resuming processing...\n{'-'*40}")

    return results
```
