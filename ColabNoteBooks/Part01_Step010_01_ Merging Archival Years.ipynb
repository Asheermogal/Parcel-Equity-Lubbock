```python
import pandas as pd
import numpy as np
```

```python
df_1945 = pd.read_excel('/content/1945_addresses_aggregated_repeated.xlsx')
df_1975 = pd.read_excel('/content/1975_addresses_aggregated_repeated.xlsx')
df_1985 = pd.read_excel('/content/1985_addresses_aggregated_repeated.xlsx')
```

```python
# Extract the address columns from each DataFrame
addresses_1945 = df_1945['google_standard_address_1945']
addresses_1975 = df_1975['google_standard_address_1975']
addresses_1985 = df_1985['google_standard_address_1985']

# Combine all addresses into one series, remove duplicates, and sort
all_addresses = pd.concat([addresses_1945, addresses_1975, addresses_1985]).drop_duplicates().sort_values()

# Create a new DataFrame from the sorted unique addresses
df = pd.DataFrame(all_addresses, columns=['address'])
```

```python
# Join df with df_1945
df_merged = pd.merge(df, df_1945, how='outer', left_on='address', right_on='google_standard_address_1945')

# Join the updated df_merged with df_1975
df_merged = pd.merge(df_merged, df_1975, how='outer', left_on='address', right_on='google_standard_address_1975')

# Join the further updated df_merged with df_1985
df_merged = pd.merge(df_merged, df_1985, how='outer', left_on='address', right_on='google_standard_address_1985')

# Display the first few rows of the final merged dataframe to verify
print(df_merged.head())

# Optionally, save the final merged dataframe to a new Excel file
df_merged.to_excel('final_merged_data.xlsx', index=False)
```
